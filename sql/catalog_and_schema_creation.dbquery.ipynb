```sql
CREATE CATALOG IF NOT EXISTS airquality
COMMENT 'Air quality monitoring data for Ontario - OpenAQ pipeline';

CREATE SCHEMA IF NOT EXISTS airquality.bronze
MANAGED LOCATION 's3://mattia-airquality-datalake-use1/bronze/'
COMMENT 'Raw data ingested from OpenAQ API - immutable';

CREATE SCHEMA IF NOT EXISTS airquality.silver
MANAGED LOCATION 's3://mattia-airquality-datalake-use1/silver/'
COMMENT 'Cleaned and validated air quality measurements';

CREATE SCHEMA IF NOT EXISTS airquality.gold
MANAGED LOCATION 's3://mattia-airquality-datalake-use1/gold/'
COMMENT 'Aggregated metrics ready for dashboards and reporting';

SHOW SCHEMAS IN airquality;
```
