```python
import requests

# API configuration
API_KEY = dbutils.secrets.get("airquality", "openaq_api_key")
BASE_URL = "https://api.openaq.org/v3"

# Database configuration
CATALOG = "airquality"
SCHEMA = "bronze"

def fetch_openaq(endpoint, params=None):
    """
    Fetch data from OpenAQ API.
    
    Args:
        endpoint: API endpoint (e.g., 'locations', 'sensors/123/latest')
        params: Optional query parameters dictionary
        
    Returns:
        JSON response as dictionary
        
    Raises:
        HTTPError: If API request fails
    """
    headers = {"X-API-Key": API_KEY}
    response = requests.get(
        f"{BASE_URL}/{endpoint}",
        headers=headers,
        params=params
    )
    response.raise_for_status()
    return response.json()
```
